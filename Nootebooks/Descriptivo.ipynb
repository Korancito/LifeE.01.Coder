¿Cómo afecta la vacunación infantil (Hepatitis B, Polio, Difteria) a la esperanza de vida y la mortalidad infantil? en LATAM

```python

def categorizar_inmunizacion(valor):
    if valor <= 40:
        return 'Baja inmunización'
    elif 40 < valor <= 70:
        return 'Inmunización media'
    elif 70 < valor <= 90:
        return 'Alta inmunización'
    else:
        return 'Inmunización completa'

dfrenamed['Hepatitis_B_Nivel'] = dfrenamed['Hepatitis B'].apply(categorizar_inmunizacion)
dfrenamed['Polio_Nivel'] = dfrenamed['Polio'].apply(categorizar_inmunizacion)
dfrenamed['Difteria_Nivel'] = dfrenamed['Difteria'].apply(categorizar_inmunizacion)

dfrenamed[['País', 'Hepatitis_B_Nivel', 'Polio_Nivel', 'Difteria_Nivel']].head()
```

```python
df_latam = dfrenamed[dfrenamed['Latam']== True]

enfermedades = ['Hepatitis B', 'Polio', 'Difteria']
niveles_inmunizacion = ['Hepatitis_B_Nivel', 'Polio_Nivel', 'Difteria_Nivel']

fig, axes = plt.subplots(1, 3, figsize=(20, 6), sharey=True)

for i, enfermedad in enumerate(enfermedades):
    sns.boxplot(x=niveles_inmunizacion[i], y='Esperanza de vida', data=df_latam, ax=axes[i], hue=niveles_inmunizacion[i])
    axes[i].set_title(f'Esperanza de Vida vs {enfermedad}')
    axes[i].set_xlabel(f'Nivel de Inmunización de {enfermedad}')
    axes[i].set_ylabel('Esperanza de Vida')

plt.tight_layout()
plt.show()
```

```python
enfermedades = ['Hepatitis B', 'Polio', 'Difteria']
niveles_inmunizacion = ['Hepatitis_B_Nivel', 'Polio_Nivel', 'Difteria_Nivel']

fig, axes = plt.subplots(1, 3, figsize=(20, 6), sharey=True)

for i, enfermedad in enumerate(enfermedades):
    sns.boxplot(x=niveles_inmunizacion[i], y='Muertes infantiles', data=df_latam, ax=axes[i], hue=niveles_inmunizacion[i])
    axes[i].set_title(f'Mortalidad infantil vs {enfermedad}')
    axes[i].set_xlabel(f'Nivel de Inmunización de {enfermedad}')
    axes[i].set_ylabel('Mortalidad Infantil')

plt.tight_layout()
plt.show()
```

Se puede observar que en funcion de una mejor inmunizacion a las diferentes enfermedades la esperanza de vida de cada nacion, es mayor, aunque tambien asi, se pueden observar algunos outliners.
Lo mismo interviene en la Mortalidad infantil, hay una relacion entre la mayor inmunizacion y la disminucion de la misma, pero se pueden observar casos extremos.
